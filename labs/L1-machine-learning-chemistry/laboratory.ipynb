## Troch formalizmu matematycznego

Wzorki matematyczne nie s kluczowe do zrozumienia przedmiotu, ale pomagaj w czytaniu publikacji, kt贸re z kolei mog sta si literatur potrzebn do realizacji projektu. Warto te偶 ustandaryzowa nazewnictwo zmiennych w kodzie.

Modele bdziemy oznacza przez $f: \mathcal{X} \to \mathcal{Y}$, a ich parametry literami greckimi, np. $\phi$. Czasami bdziemy pisa $f_\phi$, aby zaznaczy, 偶e model $f$ zale偶y od parametr贸w $\phi$.

Dla przykadu w uczeniu nadzorowanym bdziemy mie zbi贸r danych $\mathcal{D}=\{(x_i, y_i): x_i \in \mathcal{X} \land y_i \in \mathcal{Y}\}$, a zadaniem bdzie znalezienie takich parametr贸w modelu $\phi$, 偶eby $f_\phi(x) \approx y$ dla dowolnego "sensownego" $x$ i jego prawdziwej etykiety $y$. $$ {\arg \min}_\phi \frac{1}{N}\sum_{i=1}^N (f_\phi(x_i) - y_i)^2 $$
